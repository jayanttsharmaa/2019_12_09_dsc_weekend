```python
n = int(input())

i = 0
a = []
while i < n:
    a.append(i**2)
    i += 1

print(a)
# n 5
# i 0   1      2         
# a [0] [0,1]  [0,1,4]
```

```python
#  WAP to find sum of no. from 1-n
n = int(input())

s = 0
i = 1

while i <= n:
    s = s + i
    i = i + 1

print(s)
#s 0 1 3 6 10
#i 1 2 3 4 5
```
